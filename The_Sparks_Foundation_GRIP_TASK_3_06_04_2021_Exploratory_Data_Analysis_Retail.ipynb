## **Observations from Bivariate and Multivariate Analysis**

```python
df1['%Profit']=(df1['Profit']/df1['Sales'])*100
df1['Discount%']=df1['Discount']*100
df1.head()
```

#### **Observation 1** 
From the graph below we can infer that, many of the office supplies are making %loss rather than %profit when a plot of it is drawn with respect to categories, for instance Appliances, Binder in Office Supplies; Book case, Tables in Furniture and Machines are loss making entity.

```python
cstb1=pd.crosstab(columns=df1['Category'],index=df1['Sub-Category'],values=df1['%Profit'],aggfunc='mean')
cstb1.plot(kind='bar',figsize=(20,10),ylabel="%Profit",xlabel="Sub-Category",title="Average %Profit for Sub-Category")
plt.show()
cstb1=cstb1.fillna(0)
cstb1
```

#### **Observation 2**
We can see when the graph between %profit and shipping mode for every segment is done, Central region is the most loss making entity, compared to any other region,so more focus has to be given to central region to recover the losses.


```python
pt1=pd.pivot_table(data=df1,index='Region',columns=['Segment','Ship Mode'],values='%Profit')
pt1.plot(kind='bar',figsize=(20,10),xlabel="Shipping Mode in every segment",ylabel="%Profit",title="Average %Profit for every shipping mode in every segment")
plt.show()
pt1
```

#### **Observation 3**
The barplot depicts that for all the segment in the central region is loss making entity, so the focus has to be given here.

```python
pt2=pd.pivot_table(data=df1,index='Region',values='%Profit',columns='Segment')
pt2.plot(kind='bar',figsize=(20,10),xlabel="Region",ylabel="%Profit",title="Average %Profit based on Region and Segment")
plt.show()
pt2
```

#### **Observation 4**
We found out the reason as to why central is having huge loss, it is because, the central region has huge discount to unload the product. The discount should be reduced else it is leading to huge losses.

```python
pt3=pd.pivot_table(data=df1,index='Region',values='Discount',columns='Segment')
pt3.plot(kind='bar',figsize=(20,10),xlabel="Region",ylabel="%Discounts",title="Average Discount% based on Region and Segment")
plt.show()
pt3
```

#### **Observation 5**
As we can observe in the plot below, overall profits are very less compared to discount and hence there is a huge loss.

```python
pt4=pd.pivot_table(data=df1,index='Region',values=['Discount%','%Profit'])
pt4.plot(kind='bar',figsize=(20,10),xlabel="Region",ylabel="%Profit and %Discount",title="Avg %Profit and %Discount as per region")
plt.show()
pt4
```

#### **Observation 6**
In central region office supplies and FUrniture are given huge discount and hence the profit comes down

```python
pt5=pd.pivot_table(data=df1,index=['Category','Region'],values=['Discount%','%Profit'])
pt5.plot(kind='bar',figsize=(20,10),xlabel="Region",ylabel="Avg %Profit and %Discount",title="%Profit and %Discount as per region")
plt.show()
pt5
```

#### **Observation 7**
we can discern that the maximum discounts are given to these three places Consumer, Corporate,Home office in central region, the profit exceeds the discount by a huge proportion and hence making a cumulative loss.
